```python
from gapminder import gapminder
import pandas as pd
import random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

gapminder.head()

GapMinder = gapminder.copy()


print("\n\n Menú principal: \n\n   ")
print("1. Guardar gapminder con el 10% de las columnas lifeExp, pop y gdpPercap con valor no asignado aleatoriamente \n")
print("2. Importar gapminder\n")
print("3. Graficar el diagrama de dispersión lifeExp vs pop.\n")
print("4. Graficar el diagrama de dispersión gdpPercap vs pop.\n")
print("5. Graficar los diagramas de cajas de la variable gdpPercap discriminados por continentes desde 1990 a 2007.\n ")
print("0. Salir \n\n")
menu = int(input())

while menu !=0:
    
    if menu == 1:
        
        print("\n Voy a guardar gapminder en la ruta definida \n")
            
        #Hallar el 10% de las columnas lifeExp, pop y gdpPercap
        
        le_10 = round(len(GapMinder['lifeExp']) * 0.10)

        pop_10 = round(len(GapMinder['pop']) * 0.10)

        gdp_10 = round(len(GapMinder['gdpPercap']) * 0.10)

        #Seleccionar indices
        ind_le = np.random.permutation(len(GapMinder['lifeExp']))
        ind_le = ind_le[:le_10]
            
        ind_pop = np.random.permutation(len(GapMinder['pop']))
        ind_pop = ind_pop[:pop_10]
            
        ind_gdp = np.random.permutation(len(GapMinder['gdpPercap']))
        ind_gdp = ind_gdp[:gdp_10]


         #Asignar NA (Valor No Asignado) al 10% de las columnas lifeExp, pop y gdpPercap


        #print(GapMinder.loc[0,'lifeExp'])

        GapMinder.loc[ind_le,'lifeExp'] = np.nan

        GapMinder.loc[ind_pop,'pop'] = np.nan

        GapMinder.loc[ind_gdp,'gdpPercap'] = np.nan
            
        

        #Exportar a xlsx

        GapMinder.to_excel("gapminder.xlsx")
            
                  
    elif menu == 2:
        
        print("\n Importando gapminder.xlsx \n")
        
        #Importar el xlsx exportado anteriormente
        
        df = pd.read_excel("gapminder.xlsx")
        print(df.head())
            
    elif menu == 3:
        
        print("\n Realizando el diagrama de dispersión lifeExp vs pop. \n")
        
        #importar gapminder.xlsx    
        df = pd.read_excel("gapminder.xlsx")
        
        #Diagrama de dispersión
        fig, ax = plt.subplots()
        ax.scatter(df["lifeExp"], df["pop"])
        plt.show()
     
    elif menu == 4:
        
        print("Realizando el diagrama de dispersión gdpPercap vs pop. ")
        
        #importar gapminder.xlsx    
        df = pd.read_excel("gapminder.xlsx")
        
        #Diagrama de dispersión
        fig, ax = plt.subplots()
        ax.scatter(df["gdpPercap"], df["pop"])
        plt.show()
            
    elif menu == 5:
            
        print("Es necesario salir del menú, mediante la opcion 0, para visualizar el diagrama de cajas")
        
        #importar gapminder.xlsx    
        df = pd.read_excel("gapminder.xlsx")
        
        #filtrar por año
        df.filter(["year"])
        df_filtrado = df[(df.year >= 1990) & (df.year <=2007)]
        df_filtrado
        
        sns.set_style('whitegrid')
        ax = sns.boxplot(x='continent',y='gdpPercap',data=df_filtrado)
        ax = sns.stripplot(x="continent",y="gdpPercap",data=df_filtrado)
        
        
            
    else:
            
        print("Por favor digita una opción correcta")
        
    print("\nMenú principal: \n\n   ")
    print("1. Guardar gapminder con el 10% de las columnas lifeExp, pop y gdpPercap con valor no asignado aleatoriamente \n")
    print("2. Importar gapminder\n")
    print("3. Graficar el diagrama de dispersión lifeExp vs pop.\n")
    print("4. Graficar el diagrama de dispersión gdpPercap vs pop.\n")
    print("5. Graficar los diagramas de cajas de la variable gdpPercap discriminados por continentes desde 1990 a 2007.\n ")
    print("0. Salir \n\n")
    menu = int(input())

```
